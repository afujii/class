```python
import math
import numpy as np
import matplotlib.pyplot as plot

def function( x ):
	return np.power( x, 4) /4.0 + 2.0* np.power(x,3) / 3.0 \
        - np.power( x,2 )/ 2.0 - 2.0 * x + 1.0

X = np.arange(-3.1 ,2.1 ,0.1)

Y = function(X)
plot.plot(X,Y)
plot.show()
```

```python
import numpy as np
import matplotlib.pyplot as plot
from scipy.optimize import minimize

def function( x ):
	return np.power( x, 4) /4.0 + 2.0* np.power(x,3) / 3.0 \
        - np.power( x,2 )/ 2.0 - 2.0 * x + 1.0

def func_dev( x ):
        return np.power( x, 3 ) + 2.0 * np.power( x, 2 ) - x - 2.0

X = np.arange(-3.1 ,2.1 ,0.1)
Y = function(X)
plot.plot(X,Y)
plot.show()

x0 = np.array( [1.3])
res = minimize(function, x0, method='Newton-CG',
               jac= func_dev,
               hess= None,
               options={'xtol': 1e-8, 'disp': True})
print(res.x)
```
