## Open Street Map のPython を利用した扱い
ライブラリとして、staticmap と　folium を紹介する。

```python
!pip install staticmap
```

```python
# 地図オブジェクトを生成
from staticmap import StaticMap
# 画像の横幅と縦幅と OpenStreetMap 標準タイルレイヤーのタイルURLを指定
map = StaticMap(800, 600, url_template='http://a.tile.openstreetmap.org/{z}/{x}/{y}.png')

# 地図を描画した Pillow の PIL.Image オブジェクトを取得
# ズームレベルと経度・緯度を指定
from matplotlib import pyplot as plt
image = map.render(zoom=18, center=[136.882090, 35.170560])
fig, ax = plt.subplots()
ax.imshow(image)
plt.show()
```

地図の選択としては、　デフォルト地図、OpenStreetMap 、国土地理院の地図などが利用できるようである。

法政大学小金井キャンパス周辺の地図を表示する。緯度経度＝35.710, 139.522

```python
import folium
# 800x600 のサイズで縮尺　15 location を中心に地図表示
folium_map = folium.Map(width=800,height=600,location=[35.710, 139.522], zoom_start=15)

# マーカー2か所の定義、法政大と江戸東京たてもの園
folium.Marker(
    location=[35.710, 139.522],
    popup='Hosei Univ. Koganei',
    icon=folium.Icon(color='red', icon='home')
).add_to(folium_map)

folium.Marker(
    location=[35.716, 139.513],
    popup='Edo Tokyo Tatemonoen',
    icon=folium.Icon(icon="bell", icon_color='#ffff00', color="blue")

).add_to(folium_map)

folium_map
```
