# ニュートン法
高校の数学を復習します。数値計算の例を学習します。


x1からx2,x3を求める手順を示し、そこから一般的な関係式について算出します
x1からx2を算出する
下の図のようにx1, x2, f(x1)で作られる直角三角形に注目します。⊿x, ⊿yはそれぞれ直角三角形の辺の長さを表します

image.png

f'(x1)はf(x1)における接線なので、傾きの定義式の関係より、x2を求めることができます。　


$ f'
(x_1)=\frac{\Delta y}{\Delta x} \\
= \frac{f(x_1)}{x_1−x_2} \\
x_1 -x_2 = \frac{f(x_1)}{f'(x_1)}  \\
\therefore x_2 = x_1 - \frac{f(x1)}{f'(x1)} \\
$


つまり、今ニュートン法をn回行って算出したx座標xnに対してニュートン法を適応して算出されるx座標xn+1には次のような関係式が成り立ちます

$
x_{n+1} = x_n -\frac{f(x_n)}{f'(x_n)}
$


計算回数nを増やしていけば行くほど、$x_n$は真の値xに近づいていきます

```python
epsilon = 0.0001
k = 3.0 #　平方根を計算したい数字
guess = k/2.0 # 最初の予想値
while abs(guess*guess - k) >= epsilon:
    guess = guess - (((guess**2) - k)/(2*guess))
print( 'Square root of', k, 'is about', guess )
```
